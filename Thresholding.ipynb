```
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
img = cv2.imread('digital.jpg')
retval, threshold = cv2.threshold(img, 190, 255, cv2.THRESH_BINARY)

cv2_imshow(img)
cv2_imshow(threshold)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
